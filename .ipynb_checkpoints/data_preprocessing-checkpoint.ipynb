```python
numerical_cols = df.select_dtypes(include=np.number).columns

std_dev_values = df[numerical_cols].std()
variance_values = df[numerical_cols].var()
min_values = df[numerical_cols].min()
max_values = df[numerical_cols].max()
range_values = max_values - min_values

dispersion_summary_df = pd.DataFrame({
    'Độ lệch chuẩn': std_dev_values,
    'Phương sai': variance_values,
    'Min': min_values,
    'Max': max_values,
    'Phạm vi': range_values
}).T # Chuyển vị để các chỉ số thống kê thành hàng

print("Bảng tóm tắt các chỉ số xu hướng phân tán:")
display(dispersion_summary_df)
```

### Xác định giá trị ngoại lai trong `df`

```python
numerical_cols = df.select_dtypes(include=np.number).columns

outlier_summary = {}

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    outlier_summary[col] = {
        'Số lượng ngoại lai': len(outliers),
        'Phần trăm ngoại lai': (len(outliers) / len(df)) * 100
    }

print("Bảng tóm tắt các giá trị ngoại lai (sử dụng phương pháp IQR):")
display(pd.DataFrame(outlier_summary).T)

# Optionally, you can also display some of the actual outlier values for a specific column
# print("\nVí dụ 5 giá trị ngoại lai đầu tiên cho cột 'Precipitation':")
# outliers_precipitation = df[(df['Precipitation'] < (df['Precipitation'].quantile(0.25) - 1.5 * (df['Precipitation'].quantile(0.75) - df['Precipitation'].quantile(0.25)))) |
#                             (df['Precipitation'] > (df['Precipitation'].quantile(0.75) + 1.5 * (df['Precipitation'].quantile(0.75) - df['Precipitation'].quantile(0.25))))]['Precipitation']
# display(outliers_precipitation.head())
```

### Kiểm tra giá trị thiếu trong `df`

```python
missing_values = df.isnull().sum()

print("Số lượng giá trị thiếu trong mỗi cột:")
display(missing_values[missing_values > 0])

if missing_values.sum() == 0:
    print("Không có giá trị thiếu nào trong DataFrame.")
else:
    print("Có giá trị thiếu trong DataFrame. Vui lòng xem bảng trên để biết chi tiết.")
```

```python
def plot_scatter(df, x_col, y_col, title, x_label, y_label):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=x_col, y=y_col, color='coral', alpha=0.6)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Visualize relationship between Temperature and Precipitation
plot_scatter(df, 'Temperature', 'Precipitation', 'Mối quan hệ giữa Nhiệt độ và Lượng mưa', 'Nhiệt độ (°C)', 'Lượng mưa (mm)')

# Visualize relationship between Temperature and Relative Humidity
plot_scatter(df, 'Temperature', 'Relative Humidity', 'Mối quan hệ giữa Nhiệt độ và Độ ẩm tương đối', 'Nhiệt độ (°C)', 'Độ ẩm tương đối (%)')

# Visualize relationship between Temperature and Wind Speed
plot_scatter(df, 'Temperature', 'Wind Speed', 'Mối quan hệ giữa Nhiệt độ và Tốc độ gió', 'Nhiệt độ (°C)', 'Tốc độ gió (m/s)')
```

## Analyze Time Series Trends

### Subtask:
Perform time series trend analysis for numerical columns in `df`. This will involve plotting 'Precipitation', 'Temperature', 'Relative Humidity', and 'Wind Speed' against 'Date' to observe their long-term trends. Include appropriate labels and titles for clarity.


```python
numerical_cols = ['Precipitation', 'Temperature', 'Relative Humidity', 'Wind Speed']

for col in numerical_cols:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df, x='Date', y=col, color='blue')
    plt.title(f'Xu hướng {col.replace("Precipitation", "Lượng mưa").replace("Temperature", "Nhiệt độ").replace("Relative Humidity", "Độ ẩm tương đối").replace("Wind Speed", "Tốc độ gió")} theo thời gian')
    plt.xlabel('Ngày')

    # Set appropriate y-axis labels
    if col == 'Precipitation':
        plt.ylabel('Lượng mưa (mm)')
    elif col == 'Temperature':
        plt.ylabel('Nhiệt độ (°C)')
    elif col == 'Relative Humidity':
        plt.ylabel('Độ ẩm tương đối (%)')
    elif col == 'Wind Speed':
        plt.ylabel('Tốc độ gió (m/s)')

    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

## Analyze Seasonality

### Subtask:
Analyze seasonal patterns in the numerical columns of `df`. This will involve extracting month and year from the 'Date' column and then visualizing average monthly or yearly values for 'Precipitation', 'Temperature', 'Relative Humidity', and 'Wind Speed' to identify seasonal variations. Use plots like box plots or line plots for seasonal averages.


```python
numerical_cols = ['Precipitation', 'Temperature', 'Relative Humidity', 'Wind Speed']
average_yearly_data = df.groupby('Year')[numerical_cols].mean().reset_index()

for col in numerical_cols:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=average_yearly_data, x='Year', y=col, marker='o', color='purple')
    plt.title(f'Xu hướng trung bình theo năm của {col.replace("Precipitation", "Lượng mưa").replace("Temperature", "Nhiệt độ").replace("Relative Humidity", "Độ ẩm tương đối").replace("Wind Speed", "Tốc độ gió")}')
    plt.xlabel('Năm')

    # Set appropriate y-axis labels
    if col == 'Precipitation':
        plt.ylabel('Lượng mưa trung bình (mm)')
    elif col == 'Temperature':
        plt.ylabel('Nhiệt độ trung bình (°C)')
    elif 'Relative Humidity' in col:
        plt.ylabel('Độ ẩm tương đối trung bình (%)')
    elif col == 'Wind Speed':
        plt.ylabel('Tốc độ gió trung bình (m/s)')

    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

### Hiển thị các điểm ngoại lai đã được phát hiện trong `df`

```python
numerical_cols = ['Precipitation', 'Temperature', 'Relative Humidity', 'Wind Speed']

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]

    if not outliers.empty:
        print(f"\nCác giá trị ngoại lai cho cột '{col.replace('Precipitation', 'Lượng mưa').replace('Temperature', 'Nhiệt độ').replace('Relative Humidity', 'Độ ẩm tương đối').replace('Wind Speed', 'Tốc độ gió')}':")
        display(outliers)
    else:
        print(f"\nKhông tìm thấy giá trị ngoại lai cho cột '{col.replace('Precipitation', 'Lượng mưa').replace('Temperature', 'Nhiệt độ').replace('Relative Humidity', 'Độ ẩm tương đối').replace('Wind Speed', 'Tốc độ gió')}' bằng phương pháp IQR.")
```

# 4. XỬ LÝ DỮ LIỆU

## 4.1. Xử lý giá trị ngoại lai (Outlier Treatment)

Sử dụng phương pháp Capping (Winsorization) để xử lý outliers:

```python
def cap_outliers(df, columns, method='iqr', lower_percentile=0.05, upper_percentile=0.95):
    """
    Xử lý outliers bằng phương pháp capping (Winsorization)
    """
    df_capped = df.copy()
    outlier_info = {}
    
    for col in columns:
        if method == 'iqr':
            Q1 = df_capped[col].quantile(0.25)
            Q3 = df_capped[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
        else:  # percentile
            lower_bound = df_capped[col].quantile(lower_percentile)
            upper_bound = df_capped[col].quantile(upper_percentile)
        
        # Đếm số outliers trước khi capping
        outliers_before = ((df_capped[col] < lower_bound) | (df_capped[col] > upper_bound)).sum()
        
        # Thực hiện capping
        df_capped[col] = df_capped[col].clip(lower=lower_bound, upper=upper_bound)
        
        outlier_info[col] = {
            'lower_bound': lower_bound,
            'upper_bound': upper_bound,
            'outliers_capped': outliers_before
        }
    
    return df_capped, outlier_info

# Các cột cần xử lý outliers (trừ Wind Speed vì không có outliers)
numerical_cols_to_cap = ['Precipitation', 'Temperature', 'Relative Humidity']

# Xử lý outliers bằng IQR method
df_processed, outlier_info = cap_outliers(df, numerical_cols_to_cap, method='iqr')

print("=" * 60)
print("THÔNG TIN XỬ LÝ OUTLIERS")
print("=" * 60)
for col, info in outlier_info.items():
    print(f"\n{col}:")
    print(f"  - Giới hạn dưới: {info['lower_bound']:.2f}")
    print(f"  - Giới hạn trên: {info['upper_bound']:.2f}")
    print(f"  - Số giá trị đã capping: {info['outliers_capped']}")

print("\n✓ Đã xử lý outliers thành công!")
```

## 4.2. Tạo đặc trưng mới (Feature Engineering)

Tạo các đặc trưng chuỗi thời gian và thống kê rolling:

```python
# Sắp xếp dữ liệu theo ngày
df_processed = df_processed.sort_values('Date').reset_index(drop=True)

# 1. Lag features (giá trị trễ 1-7 ngày)
features_to_lag = ['Precipitation', 'Temperature', 'Relative Humidity', 'Wind Speed']
lag_days = [1, 2, 3, 7]

for col in features_to_lag:
    for lag in lag_days:
        df_processed[f'{col}_lag{lag}'] = df_processed[col].shift(lag)

# 2. Rolling statistics (trung bình, min, max 7 và 30 ngày)
for col in features_to_lag:
    # 7 ngày
    df_processed[f'{col}_roll7_mean'] = df_processed[col].rolling(window=7).mean()
    df_processed[f'{col}_roll7_min'] = df_processed[col].rolling(window=7).min()
    df_processed[f'{col}_roll7_max'] = df_processed[col].rolling(window=7).max()
    df_processed[f'{col}_roll7_std'] = df_processed[col].rolling(window=7).std()
    
    # 30 ngày
    df_processed[f'{col}_roll30_mean'] = df_processed[col].rolling(window=30).mean()
    df_processed[f'{col}_roll30_std'] = df_processed[col].rolling(window=30).std()

# 3. Đặc trưng thời gian
df_processed['DayOfYear'] = df_processed['Date'].dt.dayofyear
df_processed['WeekOfYear'] = df_processed['Date'].dt.isocalendar().week
df_processed['Quarter'] = df_processed['Date'].dt.quarter

# 4. Biến đổi cyclic cho thời gian (sin/cos)
df_processed['Month_sin'] = np.sin(2 * np.pi * df_processed['Month'] / 12)
df_processed['Month_cos'] = np.cos(2 * np.pi * df_processed['Month'] / 12)
df_processed['DayOfYear_sin'] = np.sin(2 * np.pi * df_processed['DayOfYear'] / 365)
df_processed['DayOfYear_cos'] = np.cos(2 * np.pi * df_processed['DayOfYear'] / 365)

print("=" * 60)
print("FEATURE ENGINEERING - CÁC ĐẶC TRƯNG ĐÃ TẠO")
print("=" * 60)
print(f"\n1. Lag features: {len(features_to_lag)} biến × {len(lag_days)} lag = {len(features_to_lag) * len(lag_days)} đặc trưng")
print(f"2. Rolling 7-day: {len(features_to_lag)} biến × 4 metrics = {len(features_to_lag) * 4} đặc trưng")
print(f"3. Rolling 30-day: {len(features_to_lag)} biến × 2 metrics = {len(features_to_lag) * 2} đặc trưng")
print(f"4. Time features: 8 đặc trưng (Month, Year, DayOfYear, WeekOfYear, Quarter, cyclic)")
print(f"\n📊 Tổng số đặc trưng ban đầu: 4")
print(f"📊 Tổng số đặc trưng sau khi tạo: {df_processed.shape[1] - 1}")  # Trừ cột Date
print("\n✓ Đã tạo đặc trưng mới thành công!")
```

## 4.3. Chuẩn hóa dữ liệu (Normalization)

Chuẩn hóa Min-Max cho các biến đầu vào:

```python
# Loại bỏ các hàng có giá trị NaN (do rolling và lag features)
df_processed_clean = df_processed.dropna().reset_index(drop=True)
print(f"Số mẫu sau khi loại bỏ NaN: {len(df_processed_clean)}")

# Các cột cần chuẩn hóa (features gốc, không bao gồm lag và rolling)
original_features = ['Precipitation', 'Temperature', 'Relative Humidity', 'Wind Speed']

# Chuẩn hóa Min-Max
scaler = MinMaxScaler()
df_normalized = df_processed_clean.copy()

# Chỉ chuẩn hóa các biến gốc (không chuẩn hóa lag, rolling, và cyclic features)
df_normalized[original_features] = scaler.fit_transform(df_processed_clean[original_features])

# Lưu scaler để sử dụng sau này
import joblib
scaler_path = os.path.join(data_dir, 'scaler.pkl')
joblib.dump(scaler, scaler_path)

print("\n" + "=" * 60)
print("THÔNG TIN CHUẨN HÓA")
print("=" * 60)
print(f"\nPhương pháp: Min-Max Scaling")
print(f"Số features đã chuẩn hóa: {len(original_features)}")
print(f"\nThông số scaler:")
for i, col in enumerate(original_features):
    print(f"  {col}: min={scaler.data_min_[i]:.4f}, max={scaler.data_max_[i]:.4f}")
print(f"\n✓ Đã chuẩn hóa và lưu scaler tại: {scaler_path}")
```

## 4.4. Lưu dữ liệu đã xử lý

```python
# Lưu dữ liệu đã xử lý
df_normalized_path = os.path.join(data_dir, 'rainfall_data_processed.csv')
df_normalized.to_csv(df_normalized_path, index=False)

print("=" * 60)
print("LƯU DỮ LIỆU ĐÃ XỬ LÝ")
print("=" * 60)
print(f"\nĐường dẫn file: {df_normalized_path}")
print(f"Kích thước: {df_normalized.shape[0]} hàng × {df_normalized.shape[1]} cột")
print(f"\nCác cột trong dữ liệu:")
for i, col in enumerate(df_normalized.columns, 1):
    print(f"  {i:2d}. {col}")
print("\n✓ Dữ liệu đã xử lý được lưu thành công!")
```
